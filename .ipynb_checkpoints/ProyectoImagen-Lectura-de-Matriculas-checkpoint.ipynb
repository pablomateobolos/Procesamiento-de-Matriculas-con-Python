## Introducción
Este cuaderno tiene como objetivo detectar la matrícula de un vehículo a partir de una imagen, leer sus caracteres y, posteriormente, determinar qué zonas de la ciudad de Valencia puede circular según las restricciones por emisiones. Se utilizarán herramientas de visión por computador con OpenCV y OCR con pytesseract.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import pytesseract
import os
```

### Primero cargamos todas las imagenes de la carpeta que vamos a utilizar a lo largo del trabajo y las mostramos en el cuaderno.

```python
# Ruta a la carpeta con las imágenes
image_dir = "./coches"
image_size = (256, 256)

# Función para cargar imágenes
def load_all_car_images():
    images = []
    valid_extensions = (".jpg", ".jpeg", ".png", ".bmp")
    filenames = []

    for filename in sorted(os.listdir(image_dir)):
        if filename.lower().endswith(valid_extensions):
            image_path = os.path.join(image_dir, filename)
            try:
                img = Image.open(image_path).resize(image_size)
                images.append(np.array(img))
                filenames.append(filename)
            except Exception as e:
                print(f"❌ Error al cargar {filename}: {e}")
    return np.array(images), filenames


# Cargar imágenes
X_cars, filenames = load_all_car_images()
n_images = len(X_cars)

# Mostrar todas las imágenes en un grid (ajustable)
cols = 5  # imágenes por fila
rows = int(np.ceil(n_images / cols))

plt.figure(figsize=(10, 5))

for i in range(n_images):
    plt.subplot(rows, cols, i + 1)
    plt.imshow(X_cars[i])
    plt.title(filenames[i], fontsize=8)
    plt.axis("off")

plt.tight_layout()
plt.show()
```

1. ``image_dir`` y ``image_size``: Se define la ruta de la carpeta que contiene las imágenes de los coches y el tamaño al que se redimensionarán las imágenes (256x256 píxeles).


2. ``load_all_car_images()``:
* Itera a través de los archivos en el **image_dir**.
* Verifica que los archivos tengan extensiones de imagen válidas.
* Abre cada imagen usando PIL.Image.open(), la redimensiona a image_size y la convierte a un array de NumPy.
* Almacena las imágenes procesadas y sus nombres de archivo.
* Devuelve un array de NumPy con todas las imágenes y una lista de sus nombres.
3. Carga y visualización:
+ Se llama a load_all_car_images() para obtener las imágenes y los nombres de archivo.
* Se calcula el número de filas y columnas para mostrar las imágenes en una cuadrícula ordenada.
* Se crea una figura con matplotlib.pyplot y se itera sobre cada imagen, mostrándola en un subgráfico con su nombre de archivo como título. plt.axis("off") desactiva los ejes para una visualización más limpia.
* plt.tight_layout() ajusta automáticamente los parámetros del subplot para que encajen en el área de la figura.
* plt.show() muestra la cuadrícula de imágenes

## Umbralización Básica
Esta sección demuestra una técnica de umbralización binaria simple para realzar las matrículas, preparando las imágenes para un mejor reconocimiento de caracteres. (visto en la practica 4) 

```python

# Ruta de la imagen
image_path = "./coches/1.jpg"

# Imagen a color (BGR → RGB)
img_bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Imagen en escala de grises
img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Aplicar umbralización binaria
retval, img_thresh = cv2.threshold(img_gray, 60, 255, cv2.THRESH_BINARY)

# Mostrar imágenes
plt.figure(figsize=[23, 5])
plt.subplot(131); plt.imshow(img_rgb);                  plt.title("Original (RGB)")
plt.subplot(132); plt.imshow(img_gray, cmap="gray");    plt.title("Escala de grises")
plt.subplot(133); plt.imshow(img_thresh, cmap="gray");  plt.title("Umbralización")

# Mostrar dimensiones
print("Dimensiones de la imagen umbralizada:", img_thresh.shape)
```

**1. Carga de la imagen: La imagen ``./coches/1.jpg`` se carga en dos formatos:**
* Color (BGR): cv2.IMREAD_COLOR carga la imagen con sus tres canales de color (Azul, Verde, Rojo). Luego, cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB) la convierte a RGB, que es el formato esperado por Matplotlib.
* Escala de grises: cv2.IMREAD_GRAYSCALE carga la imagen como un solo canal de intensidad.


**2. Umbralización binaria:**
* cv2.threshold(img_gray, 60, 255, cv2.THRESH_BINARY) aplica una umbralización binaria a la imagen en escala de grises (img_gray).
* Los píxeles con un valor de intensidad mayor a 60 se establecen en 255 (blanco).
* Los píxeles con un valor de intensidad menor o igual a 60 se establecen en 0 (negro).
* retval es el umbral que se utilizó (en este caso, 60), e img_thresh es la imagen umbralizada resultante.


**3. Visualización: Se muestran las tres versiones de la imagen:**
* La original en RGB.
* La imagen en escala de grises.
* La imagen umbralizada, mostrando el contraste entre el texto y el fondo.
* Dimensiones: Se imprime la forma de la imagen umbralizada para verificar sus dimensiones.

**ELIMINAR**
* Este código carga una imagen desde la carpeta `./coches/1.jpg` y la transforma de color a escala de grises. 
* Luego, aplica una **umbralización binaria**: los píxeles con intensidad mayor a 60 se vuelven blancos (255), y los demás, negros (0). 
* Este proceso permite **separar claramente el objeto del fondo**, y mostrar el proceso al paso de las tres versiones: la imagen original, en grises y umbralizada, para comparar los efectos visualmente.

## Umbralización Adaptativa
Por lo que vemos, la matrícula no se muestra con tanta nitidez como nos gustaría. Por ello vamos a seguir aplicando la umbralización.
Pero esta vez se trata de una umbralización adaptativa

```python
# Leer la imagen en escala de grises desde la carpeta "coches"
img_read = cv2.imread("./coches/1.jpg", cv2.IMREAD_GRAYSCALE)

# Umbralización global con umbral = 50
retval, img_thresh_gbl_1 = cv2.threshold(img_read, 50, 255, cv2.THRESH_BINARY)

# Umbralización global con umbral = 130
retval, img_thresh_gbl_2 = cv2.threshold(img_read, 130, 255, cv2.THRESH_BINARY)

# Umbralización adaptativa (media local)
img_thresh_adp = cv2.adaptiveThreshold(img_read, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 7)

# Mostrar imágenes en una cuadrícula
plt.figure(figsize=[10, 5])
plt.subplot(221); plt.imshow(img_read,        cmap="gray"); plt.title("Original")
plt.subplot(222); plt.imshow(img_thresh_gbl_1, cmap="gray"); plt.title("Umbral global: 50")
plt.subplot(223); plt.imshow(img_thresh_gbl_2, cmap="gray"); plt.title("Umbral global: 130")
plt.subplot(224); plt.imshow(img_thresh_adp,   cmap="gray"); plt.title("Umbral adaptativo")

plt.tight_layout()
plt.show()
```

1. Carga de la imagen: Se carga la imagen ./coches/1.jpg en escala de grises.


2. Umbralizaciones globales (comparación): Se aplican dos umbralizaciones binarias globales con diferentes umbrales (50 y 130) para ilustrar cómo un umbral fijo puede no ser óptimo para toda la imagen si hay variaciones de iluminación.


3. Umbralización adaptativa:
cv2.adaptiveThreshold() calcula el umbral para cada pequeño vecindario de píxeles, lo que lo hace muy efectivo para imágenes con iluminación variable.
img_read: La imagen de entrada en escala de grises.
255: El valor no cero que se asigna a los píxeles que exceden el umbral.
cv2.ADAPTIVE_THRESH_MEAN_C: El método de umbralización adaptativa. Calcula el umbral como la media del área del vecindario menos un valor constante C.
cv2.THRESH_BINARY: Tipo de umbralización binaria.
15: blockSize, el tamaño del vecindario (área) sobre el que se calcula el umbral. Debe ser un número impar.
7: C, una constante que se resta de la media o media ponderada. Esto ayuda a ajustar la sensibilidad del umbral.


4. Visualización: Se muestran la imagen original en escala de grises, las dos umbralizaciones globales y la umbralización adaptativa, lo que permite una comparación visual directa de los resultados. La umbralización adaptativa generalmente produce mejores resultados para matrículas debido a su capacidad para manejar gradientes de iluminación.

## Detección y Lectura de Matrículas con Umbralización Adaptativa y EasyOCR
Esta es la sección central del código, donde se combina la umbralización adaptativa con EasyOCR para detectar y leer el texto de las matrículas en las imágenes.

```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import os

# Crear lector con idioma español
reader = easyocr.Reader(['es'], gpu=False) # Puedes poner gpu=True si tienes GPU y drivers

# Ruta de la carpeta
carpeta = "coches"

if not os.path.exists(carpeta):
    os.makedirs(carpeta)
    print(f"Carpeta '{carpeta}' creada. Por favor, añade imágenes de coches en ella.")

archivos = os.listdir(carpeta)
archivos_imagen = [f for f in archivos if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

if not archivos_imagen:
    print(f"No se encontraron imágenes en la carpeta '{carpeta}'.")
    exit()
print("Dentro de la carpeta se encuentran los archivos:", ', '.join(archivos_imagen))

nombre_archivo = input(f"Introduce el nombre del archivo (por ejemplo: {archivos_imagen[0] if archivos_imagen else '1.jpg'}): ")
ruta_imagen = os.path.join(carpeta, nombre_archivo)

if not os.path.exists(ruta_imagen):
    print(f"❌ La imagen '{ruta_imagen}' no existe. Verifica que esté en la carpeta '{carpeta}'.")
    exit()

# 1. Leer la imagen original (a color, solo para obtener los datos iniciales)
original_image_color = cv2.imread(ruta_imagen)
if original_image_color is None:
    print(f"❌ No se pudo cargar la imagen: {ruta_imagen}. Puede estar corrupta o no ser un formato válido.")
    exit()

# --- PREPROCESAMIENTO: UMBRAL ADAPTATIVO ---
# 2. Convertir la imagen original a escala de grises
gray_image = cv2.cvtColor(original_image_color, cv2.COLOR_BGR2GRAY)

# 3. Aplicar umbral adaptativo
block_size = 11  # Ajusta según sea necesario (debe ser impar)
C_value = 5      # Ajusta según sea necesario
image_umbralizada = cv2.adaptiveThreshold(
    gray_image,
    255,  # Valor máximo a asignar
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY,
    block_size,
    C_value
)
# 'image_umbralizada' es ahora una imagen binaria (blanco y negro)

# --- LECTURA DE TEXTO CON EASYOCR (SOBRE LA IMAGEN UMBRALIZADA) ---
print("\nProcesando imagen con EasyOCR (sobre la imagen umbralizada)...")
# 4. EasyOCR lee el texto directamente de la imagen umbralizada
results = reader.readtext(image_umbralizada)

# --- DIBUJAR Y MOSTRAR RESULTADOS SOBRE LA IMAGEN UMBRALIZADA ---
if not results:
    print("No se detectó texto en la imagen umbralizada.")
    # Opcional: Mostrar la imagen umbralizada sin detecciones
    plt.figure(figsize=(10, 5))
    plt.imshow(image_umbralizada, cmap='gray') # Mostrar en escala de grises
    plt.title("Imagen Umbralizada - Sin texto detectado")
    plt.axis('off')
    plt.show()
else:
    print("\nTexto detectado:")
    # Para dibujar rectángulos y texto de colores sobre la imagen umbralizada (que es monocromática),
    # necesitamos convertirla primero a un formato de 3 canales (BGR).
    # Las coordenadas de 'results' son relativas a 'image_umbralizada'.
    image_umbralizada_para_dibujar_bgr = cv2.cvtColor(image_umbralizada, cv2.COLOR_GRAY2BGR)

    for (bbox, text, prob) in results:
        (tl, tr, br, bl) = bbox
        tl = tuple(map(int, tl))
        br = tuple(map(int, br))

        # Dibujar el rectángulo sobre la imagen umbralizada (convertida a BGR)
        cv2.rectangle(image_umbralizada_para_dibujar_bgr, tl, br, (0, 255, 0), 2) # Rectángulo verde
        # Poner el texto detectado sobre la imagen umbralizada (convertida a BGR)
        cv2.putText(image_umbralizada_para_dibujar_bgr, text, (tl[0], tl[1] - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2, cv2.LINE_AA) # Texto azul

        print(f"  Texto: '{text}', Confianza: {prob:.2f}")

    # Mostrar la imagen UMBRALIZADA con las detecciones
    # Convertir de BGR (usado por OpenCV para dibujar) a RGB (usado por Matplotlib para mostrar)
    image_umbralizada_con_detecciones_rgb = cv2.cvtColor(image_umbralizada_para_dibujar_bgr, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(12, 7))
    plt.imshow(image_umbralizada_con_detecciones_rgb)
    plt.axis('off')
    plt.title("Texto detectado sobre la Imagen Umbralizada")
    plt.show()

print("\nProceso completado.")
```

1. Inicialización de EasyOCR:
reader = easyocr.Reader(['es'], gpu=False) crea una instancia del lector de EasyOCR.
['es'] especifica que el lector debe ser entrenado para el idioma español.
gpu=False indica que se utilizará la CPU. Si se tiene una GPU compatible y los drivers instalados, se puede cambiar a True para un procesamiento más rápido.


2. Gestión de archivos y entrada del usuario:
El código verifica si la carpeta "coches" existe y, si no, la crea, pidiendo al usuario que coloque imágenes allí.
Lista los archivos de imagen válidos en la carpeta y le pide al usuario que introduzca el nombre de la imagen que desea procesar.
Verifica si la imagen introducida por el usuario existe.


3. Carga de la imagen original:
original_image_color = cv2.imread(ruta_imagen) carga la imagen seleccionada por el usuario en formato de color (BGR).
Se incluye una verificación para asegurar que la imagen se cargó correctamente.


4. Preprocesamiento: Umbralización adaptativa:
gray_image = cv2.cvtColor(original_image_color, cv2.COLOR_BGR2GRAY) convierte la imagen a escala de grises, un paso necesario para la umbralización.
cv2.adaptiveThreshold() se aplica a la imagen en escala de grises. Los parámetros block_size (tamaño del vecindario, debe ser impar) y C_value (constante a restar de la media) pueden ajustarse para optimizar la detección en diferentes condiciones de imagen.
El resultado, image_umbralizada, es una imagen binaria (blanco y negro), donde los caracteres de la matrícula deberían destacarse claramente.


5. Lectura de texto con EasyOCR:
results = reader.readtext(image_umbralizada) es la función clave. EasyOCR toma la imagen umbralizada como entrada y devuelve una lista de tuplas.
Cada tupla (bbox, text, prob) contiene:
bbox: Las coordenadas de la caja delimitadora (bounding box) del texto detectado.
text: El texto reconocido.
prob: La probabilidad de confianza del reconocimiento.


6. Dibujo y visualización de resultados:
Se verifica si se detectó texto. Si no, se muestra la imagen umbralizada sin detecciones.
Si se detecta texto, la image_umbralizada (que es monocromática) se convierte a BGR (cv2.cvtColor(image_umbralizada, cv2.COLOR_GRAY2BGR)) para poder dibujar rectángulos y texto de colores sobre ella.
Para cada resultado (bbox, text, prob):
Se extraen las coordenadas de la caja delimitadora (tl para top-left, br para bottom-right).
cv2.rectangle() dibuja un rectángulo verde alrededor del texto detectado.
cv2.putText() superpone el texto reconocido en azul encima del rectángulo, junto con su puntuación de confianza.
Finalmente, la imagen resultante con las detecciones se convierte de BGR a RGB para ser mostrada correctamente por matplotlib.pyplot.
plt.imshow() y plt.show() muestran la imagen umbralizada con los rectángulos y el texto identificados.
