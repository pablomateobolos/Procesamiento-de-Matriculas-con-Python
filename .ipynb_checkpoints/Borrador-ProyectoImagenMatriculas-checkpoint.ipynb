## 1. Cargar una Imagen de un Coche desde el Sistema de Archivos (modo local)

Cargamos las librerías que usaremos a lo largo del cuaderno

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import pytesseract
import os
```

### Primero cargamos todas las imagenes de la carpeta que vamos a utilizar a lo largo del trabajo y las mostramos en el cuaderno.

```python
# Ruta a la carpeta con las imágenes
image_dir = "./coches"
image_size = (128, 128)

# Función para cargar imágenes
def load_all_car_images():
    images = []
    valid_extensions = (".jpg", ".jpeg", ".png", ".bmp")
    filenames = []

    for filename in sorted(os.listdir(image_dir)):
        if filename.lower().endswith(valid_extensions):
            image_path = os.path.join(image_dir, filename)
            try:
                img = Image.open(image_path).resize(image_size)
                images.append(np.array(img))
                filenames.append(filename)
            except Exception as e:
                print(f"❌ Error al cargar {filename}: {e}")
    return np.array(images), filenames


# Cargar imágenes
X_cars, filenames = load_all_car_images()
n_images = len(X_cars)

# Mostrar todas las imágenes en un grid (ajustable)
cols = 4  # imágenes por fila
rows = int(np.ceil(n_images / cols))

plt.figure(figsize=(10, 5))

for i in range(n_images):
    plt.subplot(rows, cols, i + 1)
    plt.imshow(X_cars[i])
    plt.title(filenames[i], fontsize=8)
    plt.axis("off")

plt.tight_layout()
plt.show()
```

Ahora vamos a probar varias tecnicas de emmascaramiento de imagenes para que se puedan leer mejor los numeros y letras de las matriculas, que es lo que nos importa del proyecto
Para ello primero vamos a aplicar una umbralizacion avanzada (visto en la practica 4) 

```python

# Ruta de la imagen
image_path = "./coches/1.jpg"

# Imagen a color (BGR → RGB)
img_bgr = cv2.imread(image_path, cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Imagen en escala de grises
img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Aplicar umbralización binaria
retval, img_thresh = cv2.threshold(img_gray, 60, 255, cv2.THRESH_BINARY)

# Mostrar imágenes
plt.figure(figsize=[23, 5])
plt.subplot(131); plt.imshow(img_rgb);                  plt.title("Original (RGB)")
plt.subplot(132); plt.imshow(img_gray, cmap="gray");    plt.title("Escala de grises")
plt.subplot(133); plt.imshow(img_thresh, cmap="gray");  plt.title("Umbralización")

# Mostrar dimensiones
print("Dimensiones de la imagen umbralizada:", img_thresh.shape)
```

* Este código carga una imagen desde la carpeta `./coches/1.jpg` y la transforma de color a escala de grises. 
* Luego, aplica una **umbralización binaria**: los píxeles con intensidad mayor a 60 se vuelven blancos (255), y los demás, negros (0). 
* Este proceso permite **separar claramente el objeto del fondo**, y mostrar el proceso al paso de las tres versiones: la imagen original, en grises y umbralizada, para comparar los efectos visualmente.


Por lo que vemos, la matrícula no se muestra con tanta nitidez como nos gustaría. Por ello vamos a seguir aplicando la umbralización.
Pero esta vez se trata de una umbralización adaptativa

```python
# Leer la imagen en escala de grises desde la carpeta "coches"
img_read = cv2.imread("./coches/1.jpg", cv2.IMREAD_GRAYSCALE)

# Umbralización global con umbral = 50
retval, img_thresh_gbl_1 = cv2.threshold(img_read, 50, 255, cv2.THRESH_BINARY)

# Umbralización global con umbral = 130
retval, img_thresh_gbl_2 = cv2.threshold(img_read, 130, 255, cv2.THRESH_BINARY)

# Umbralización adaptativa (media local)
img_thresh_adp = cv2.adaptiveThreshold(img_read, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 7)

# Mostrar imágenes en una cuadrícula
plt.figure(figsize=[10, 5])
plt.subplot(221); plt.imshow(img_read,        cmap="gray"); plt.title("Original")
plt.subplot(222); plt.imshow(img_thresh_gbl_1, cmap="gray"); plt.title("Umbral global: 50")
plt.subplot(223); plt.imshow(img_thresh_gbl_2, cmap="gray"); plt.title("Umbral global: 130")
plt.subplot(224); plt.imshow(img_thresh_adp,   cmap="gray"); plt.title("Umbral adaptativo")

plt.tight_layout()
plt.show()
```

Cambiando el umbral ya podemos ver con mayor resolucion la matricula

```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import os

# Crear lector con idioma español
reader = easyocr.Reader(['es'], gpu=False)

# Solicitar al usuario el nombre del archivo
nombre_archivo = input("Introduce el nombre del archivo (por ejemplo: 1.jpg): ")
ruta_imagen = os.path.join("coches", nombre_archivo)

# Verificar si la imagen existe
if not os.path.exists(ruta_imagen):
    print("❌ La imagen no existe. Verifica que esté en la carpeta 'coches'.")
    exit()

# Leer imagen
image = cv2.imread(ruta_imagen)
if image is None:
    print("❌ No se pudo leer la imagen.")
    exit()

# Leer texto con EasyOCR
results = reader.readtext(image)

# Dibujar resultados
for (bbox, text, prob) in results:
    # Coordenadas del bounding box
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = tuple(map(int, top_left))
    bottom_right = tuple(map(int, bottom_right))

    # Dibujar recuadro y texto
    cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)
    cv2.putText(image, text, (top_left[0], top_left[1] - 10), cv2.FONT_HERSHEY_SIMPLEX,
                0.8, (0, 0, 255), 2, cv2.LINE_AA)

    print(f"[{prob:.2f}] {text}")

# Convertir a RGB para mostrar con matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Mostrar resultado
plt.figure(figsize=(20, 10))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Texto detectado con EasyOCR")
plt.show()
```

```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import os

# Crear lector con idioma español
reader = easyocr.Reader(['es'], gpu=False)

# Ruta de la carpeta
carpeta = "coches"

# Listar archivos en la carpeta
archivos = os.listdir(carpeta)
archivos_imagen = [f for f in archivos if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

# Mostrar los archivos encontrados
print("Dentro de la carpeta se encuentran los archivos:", ', '.join(archivos_imagen))

# Solicitar al usuario el nombre del archivo
nombre_archivo = input("Introduce el nombre del archivo (por ejemplo: 1.jpg): ")
ruta_imagen = os.path.join(carpeta, nombre_archivo)

# Validar existencia
if not os.path.exists(ruta_imagen):
    print("❌ La imagen no existe. Verifica que esté en la carpeta 'coches'.")
    exit()

# Leer imagen
image = cv2.imread(ruta_imagen)

# Leer texto con EasyOCR
results = reader.readtext(image)

# Dibujar resultados
for (bbox, text, prob) in results:
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = tuple(map(int, top_left))
    bottom_right = tuple(map(int, bottom_right))

    cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)
    cv2.putText(image, text, (top_left[0], top_left[1] - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2, cv2.LINE_AA)

    print(f"[{prob:.2f}] {text}")

# Mostrar imagen resultante
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Texto detectado con EasyOCR")
plt.show()
```

```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import os
import numpy as np # Necesario para adaptiveThreshold si no está ya importado

# Crear lector con idioma español
# Inicializar el lector una sola vez para mejorar el rendimiento si se procesan múltiples imágenes
# Descargará el modelo la primera vez que se ejecute.
print("Inicializando EasyOCR Reader (puede tardar un momento la primera vez)...")
try:
    reader = easyocr.Reader(['es'], gpu=False) # 'es' para español
    print("EasyOCR Reader inicializado.")
except Exception as e:
    print(f"Error inicializando EasyOCR: {e}")
    print("Asegúrate de tener instalados los modelos de lenguaje necesarios.")
    exit()

# Solicitar al usuario el nombre del archivo
nombre_archivo = input("Introduce el nombre del archivo de la carpeta 'coches' (por ejemplo: 1.jpg): ")
ruta_imagen = os.path.join("coches", nombre_archivo)

# Verificar si la imagen existe
if not os.path.exists(ruta_imagen):
    print(f"❌ La imagen no existe en la ruta: {ruta_imagen}. Verifica que esté en la carpeta 'coches'.")
    exit()

# 1. Leer imagen original
image_original_bgr = cv2.imread(ruta_imagen)
if image_original_bgr is None:
    print("❌ No se pudo leer la imagen.")
    exit()

# Convertir original a RGB para mostrar con matplotlib
image_original_rgb = cv2.cvtColor(image_original_bgr, cv2.COLOR_BGR2RGB)

# 2. Generar imagen con umbralización adaptativa
# Convertir a escala de grises primero
img_gray = cv2.cvtColor(image_original_bgr, cv2.COLOR_BGR2GRAY)

# Aplicar umbralización adaptativa
# Parámetros (imagen, valor máximo, método adaptativo, tipo de umbral, tamaño del bloque, constante C)
# Estos parámetros (15, 7) son los que usaste antes, puedes ajustarlos si es necesario
img_thresh_adp = cv2.adaptiveThreshold(
    img_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 7
)

# 3. Aplicar OCR a la imagen de umbralización generada
print(f"\nProcesando texto en la imagen umbralizada '{nombre_archivo}' con EasyOCR...")
# EasyOCR puede funcionar directamente con la imagen en escala de grises (o BGR)
# Si la umbralización ayuda a EasyOCR, es bueno. A veces EasyOCR prefiere la imagen original o en grises.
# Vamos a probar con la imagen umbralizada como solicitaste.
results = reader.readtext(img_thresh_adp)
print("Resultados del OCR:")

# 4. Dibujar resultados del OCR en una copia de la imagen umbralizada
# Convertimos la imagen umbralizada (que es de 1 canal) a BGR (3 canales) para poder dibujar en color
img_thresh_adp_bgr_for_drawing = cv2.cvtColor(img_thresh_adp, cv2.COLOR_GRAY2BGR)

if not results:
    print("No se detectó texto en la imagen umbralizada.")
else:
    for (bbox, text, prob) in results:
        # Coordenadas del bounding box
        (top_left, top_right, bottom_right, bottom_left) = bbox
        top_left = tuple(map(int, top_left))
        bottom_right = tuple(map(int, bottom_right))

        # Dibujar recuadro y texto en la copia BGR de la imagen umbralizada
        cv2.rectangle(img_thresh_adp_bgr_for_drawing, top_left, bottom_right, (0, 255, 0), 2) # Verde
        cv2.putText(img_thresh_adp_bgr_for_drawing, text, (top_left[0], top_left[1] - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2, cv2.LINE_AA) # Azul

        print(f"[{prob:.2f}] {text}")


# Convertir la imagen umbralizada con dibujos a RGB para mostrar con matplotlib
img_thresh_adp_rgb_with_ocr = cv2.cvtColor(img_thresh_adp_bgr_for_drawing, cv2.COLOR_BGR2RGB)


# 5. Mostrar imágenes en subplots
plt.figure(figsize=(20, 10)) # Ajusta el tamaño según necesites

# Subplot 1: Imagen Original
plt.subplot(1, 2, 1)
plt.imshow(image_original_rgb)
plt.title("Imagen Original")
plt.axis('off')

# Subplot 2: Imagen con Umbralización Adaptativa y OCR
plt.subplot(1, 2, 2)
plt.imshow(img_thresh_adp_rgb_with_ocr)
plt.title("Umbralización Adaptativa con OCR")
plt.axis('off')

plt.tight_layout() # Ajusta los subplots para que no se superpongan
plt.show()

print("\nProceso completado.")
```
